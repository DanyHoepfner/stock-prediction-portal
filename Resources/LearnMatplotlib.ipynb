```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Line Chart

```python
x = [0, 1, 2, 3, 4, 5]
y = [4, 6, 2, 7, 9, 1]
```

```python
plt.plot(x, y);
plt.title("Simple Line Plot");
plt.xlabel("x-axis");
plt.ylabel("y-axis");
plt.grid(True);
```

# Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr' ]
revenue = [359, 122, 400, 245]
```

```python
plt.bar(months, revenue);
plt.title("Simple Bar Chart");
plt.xlabel("months");
plt.ylabel("Revenue");
plt.grid(True);
```

## Subplots and Legends 

```python
days = [1,2,3,4,5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, "b-o", label="Original Price");
plt.plot(days, predicted_price, "r-o", label="Predicted Price");
plt.legend()
plt.xlabel("days")
plt.ylabel("Price")
plt.title("Plot with legends")
```

# Subplots 

```python
plt.subplot(3, 1, 1)
plt.plot(days, original_price, "b-o", label="Original Price");
plt.legend()
plt.xlabel("days")
plt.ylabel("Price")
plt.title("Subplot 1");

plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, "r-o", label="Predicted Price");
plt.legend()
plt.xlabel("days")
plt.ylabel("Price")
plt.title("Subplot 2");



# price difference

price_difference = np.array(original_price) - np.array(predicted_price)

plt.subplot(3, 1, 3)
plt.bar(days, price_difference, color="purple", label="Price Difference");
plt.legend()
plt.xlabel("days")
plt.ylabel("Price")
plt.title("Price Difference");
```

```python
plt.plot(days, original_price, "b-o", label="Original Price");
plt.plot(days, predicted_price, "r-o", label="Predicted Price");
plt.legend()
plt.xlabel("days")
plt.ylabel("Price")
plt.title("Plot with legends")
plt.savefig("example_prediction.png", dpi=300) # save the image
```
